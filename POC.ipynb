```python
v = Validator()
v.schema = {'Period': { 'type': 'float'},
'Inst_sector_code': {'type': 'integer'}}
```

Checking added data is valid or not

```python
if v.validate({'Period': 2007.03, 'Inst_sector_code': 16}):
   print('Data is valid')
else:
   print('Data is invalid')
   print(v.errors)
```

Checking whole column whether it is valid or not

```python
flag = True
for p,q in zip(df["Period"],df["Inst_sector_code"]):
    if v.validate({'Period':p,'Inst_sector_code':q}):
        continue
    else:
        flag=False
        print('Data is invalid')
        print(v.errors)
        break
if flag==True:
    print('Data is valid')
```
